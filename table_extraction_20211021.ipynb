# Import Libraries

```python
import camelot
import pandas as pd
import numpy as np
from IPython.display import display
import os
from pathlib import Path
import ocrmypdf
```

# Paths

```python
input_path_to_a = Path('data/a.pdf')
input_path_to_b = Path('data/b.pdf')
process_pdf_path_a = Path('process_data/a.pdf')
process_pdf_path_b = Path('process_data/b.pdf')
output_path_a = Path('results/a.xlsx')
output_path_b = Path('results/b.xlsx')
```

# PDF A

## OCR 

```python
ocrmypdf.ocr(str(input_path_to_a), str(process_pdf_path_a), deskew=True)
```

## Table Extraction with Lattice

```python
tables_a_using_default_lattice = camelot.read_pdf(str(process_pdf_path_a),
                                               flavor='lattice',
#                                                 line_scale=60,
#                                                 edge_tol=500,
#                                                   backend="ghostscript",
                                               pages='1')
```

```python
writer = pd.ExcelWriter(output_path_a)

for table in tables_a_using_default_lattice:
    print(table.page)
#     sheet_name = 'page_'+str(table.page)
#     table.df.to_excel(writer, sheet_name = sheet_name)

# writer.save()
# writer.close()  
```

```python
table.df
```

```python
camelot.plot(table, kind='line').show()
```
